```python
# 5)
import numpy as np
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
X = np.array([[1, 9, 5],[3, 3, 5],[0.7, 0.9, 0.2]])
w = np.array([1, 1, 1])
```

```python
lambda_H = 1
sig = sigmoid(w.T @ X)
weights = sig * (1 - sig)
I = np.eye(3,3)
H = (X * weights) @ X.T + (1/lambda_H)*I
print(H)
```

```python
# 9)
import numpy as np
X = np.array([[1,9,5],[3,3,5],[0.7,0.9,0.2]])
w = np.array([1,1,1])
y = np.array([0,1,0])
```

```python
lambda_m = 1
# m=∇l=−∑n=1N(yn−σ(wTxn))xn+1λw
sig = sigmoid(w.T @ X)
weights = y - sig
gradient_data = -(X @ weights)
m = gradient_data + (1 / lambda_m) * w
print(m)
```
